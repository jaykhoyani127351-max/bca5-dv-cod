```python
import matplotlib.pyplot as plt

year = [2016, 2017, 2018, 2019, 2020]
profit = [80000, 110000, 90000, 85000, 120000]
sales = [80000, 120000, 65000, 70000, 125000]

plt.plot(year,profit,color='red',label = 'Year vs Profit')
plt.plot(year,sales,color='green',label = 'Year vs Sales')

plt.xlabel('Year')
plt.ylabel('Sales/Profit')
plt.legend()
```
